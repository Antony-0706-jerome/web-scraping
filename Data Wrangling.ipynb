```python
# Creating Data Frame

df = pd.DataFrame()
reg_no , name , total = [] , [] , []
```

```python
inputs = int(input("Number of inputs to be entered : "))
for i in range(inputs):
    reg_no.append(int(input("Enter the Reg : ")))
    name.append(input("Enter the name : "))
    total.append(int(input("Enter the total : ")))
```

```python
df['Reg No'] , df['Name'] , df['Total'] = reg_no , name , total
df
```

```python
# Adding new row to the existing dataframe

new_entry = pd.Series([1232 , 'Suki' , '550'] , index=['Reg No' , 'Name' , 'Total'])
df.append(new_entry , ignore_index=True)
```

```python
# Concatinating the dataframe

data_a = {"id":[1 , 2 , 3] , "name":["Antony" , "Jerome" , "suki"]}
data_b = {"id":[4 , 5 , 6] , "name" : ['Jack' , 'Raj' , 'kumar']}
df1 = pd.DataFrame(data=data_a , columns=['id' , 'name'])
df2 = pd.DataFrame(data=data_b , columns=['id' , 'name'])
```

```python
# Concatinate on row wise

pd.concat([df1 , df2] , axis=0)
```

```python
# Concatinate on column wise

pd.concat([df1 , df2] , axis=1)
```
